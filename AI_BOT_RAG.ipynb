```python
def upsert_to_pinecone(ids, vectors, metadata):
    # Create upsert tuples with IDs, vectors, and metadata
    upserts = [
        (ids[i], vectors[i], {"text": metadata[i]})
        for i in range(len(ids))
    ]
    index.upsert(vectors=upserts)

ids = [str(i) for i in range(len(documents))]
# Insert the embeddings along with metadata (document text) into Pinecone
upsert_to_pinecone(ids, embeddings, documents)
```

# Building the RAG model

```python
def gen_context(query):
  embed = get_embeddings([query])
  res = index.query(vector=embed.data[0].embedding, top_k=3, include_metadata=True)
  res = res['matches'][0]
  #print(res)
  contexts = res['metadata']['text']
  return contexts

def gen_prompt(query, contexts):
  prompt_start = (
      "Answer the question based on the context below. Be creative, business minded, business friendly, diplomatic, fun and interactive \n\n" +
      "Context:\n"
  )

  prompt_end = (
      f"\n\nQuestion: {query}\nAnswer:"
  )

  prompt = (
      prompt_start + (contexts) +
      prompt_end
  )
  return prompt

def query_embedding(query):
    response = client.embeddings.create(input=[query], model="text-embedding-ada-002")
    return response.data[0].embedding


def query_pinecone(query):
    query_vec = query_embedding(query)
    result = index.query(vector=embed.data[0].embedding, top_k=3, include_metadata=True)
    #result = index.query(vector=query_vec, top_k=3, namespace='my_namespace', include_metadata=True)  # Adjust top_k as needed
    return result

#user_query = "contact?"
#results = query_pinecone(user_query)
```

```python
def generate_response(query, prompt):
    response = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prompt,
    temperature=0.8,
    max_tokens=1000,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
    stop=None)


    print('-' * 80)
    print(response.choices[0].text)
```
